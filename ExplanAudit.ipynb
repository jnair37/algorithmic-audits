```python
# Boilerplate check: Hugging Face login + LLaMA access

# Fix for numpy.dtype size changed error
# This error often occurs due to binary incompatibility between numpy and other libraries.
# Reinstalling numpy and transformers can resolve it.

from huggingface_hub import whoami, HfApi
from transformers import AutoTokenizer

# 1. Check Hugging Face login
try:
    user_info = whoami()
    print(f"✅ Logged into Hugging Face as: {user_info['name']}")
except Exception as e:
    raise RuntimeError(
        "❌ Not logged into Hugging Face.\n"
        "Run `from huggingface_hub import login; login()` and paste your token."
    )

# 2. Check access to Meta LLaMA (license-gated)
LLAMA_MODEL_ID = "meta-llama/Llama-2-7b-hf"  # change if you want a different size/version

try:
    tokenizer = AutoTokenizer.from_pretrained(LLAMA_MODEL_ID)
    print(f"✅ Access confirmed for model: {LLAMA_MODEL_ID}")
except Exception as e:
    raise RuntimeError(
        f"❌ No access to {LLAMA_MODEL_ID}.\n"
        "Make sure:\n"
        "1) You requested and were granted access on the model page\n"
        "2) You accepted Meta's license\n"
        "3) You're logged in with the correct Hugging Face account"
    )
```

```python
from transformers import pipeline
from typing import List, Dict, Any
import json
import torch

def cycle_models(model_names: List[str], input_text: str, max_tokens: int = 100) -> Dict[str, str]:
    """
    Cycles through a list of HuggingFace language models and generates outputs.

    Args:
        model_names: List of HuggingFace model identifiers (e.g., ['gpt2', 'distilgpt2'])
        input_text: The input text prompt to send to each model
        max_tokens: Maximum number of tokens to generate (default: 100)

    Returns:
        Dictionary mapping model names to their generated outputs
    """
    results = {}

    for model_name in model_names:
        try:
            # Clear CUDA cache before loading each model
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

            # Create a text-generation pipeline for each model
            generator = pipeline(
                'text-generation',
                model=model_name,
                device_map='auto' if torch.cuda.is_available() else None,
                torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
            )

            # Generate text with the model
            output = generator(
                input_text,
                max_new_tokens=max_tokens,  # Limit generation to specified tokens
                num_return_sequences=1,
                truncation=True,
                return_full_text=False,  # Only return newly generated text
                pad_token_id=generator.tokenizer.eos_token_id,  # Prevent padding issues
                do_sample=True,  # Use sampling to avoid greedy decoding issues
                temperature=0.7
            )

            # Extract the generated text (only new tokens, not including prompt)
            generated_text = output[0]['generated_text']
            results[model_name] = generated_text

            print(f"✓ Completed: {model_name}")

            # Clean up to prevent memory issues
            del generator
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

        except Exception as e:
            # Handle errors gracefully and continue with other models
            error_msg = str(e)
            # For CUDA errors, try to get more detail
            if torch.cuda.is_available() and "CUDA" in error_msg:
                torch.cuda.synchronize()  # Force CUDA to finish and reveal the real error
            results[model_name] = f"Error: {error_msg}"
            print(f"✗ Failed: {model_name} - {error_msg}")

    return results


def sweep_experiments(
    model_names: List[str],
    input_texts: List[str],
    token_lengths: List[int]
) -> Dict[str, Any]:
    """
    Sweeps over multiple inputs and token lengths, running all models on each combination.
    Loads each model once and tests all input/token combinations before moving to next model.

    Args:
        model_names: List of HuggingFace model identifiers
        input_texts: List of input text prompts to test
        token_lengths: List of token length limits to test

    Returns:
        Nested dictionary structure:
        {
            'input_0': {
                'text': 'the input text',
                'experiments': {
                    'token_length_50': {
                        'model1': 'output',
                        'model2': 'output'
                    },
                    'token_length_100': {...}
                }
            },
            'input_1': {...}
        }
    """
    # Initialize results structure
    results = {}
    for i, input_text in enumerate(input_texts):
        input_key = f"input_{i}"
        results[input_key] = {
            'text': input_text,
            'experiments': {}
        }
        for token_length in token_lengths:
            token_key = f"token_length_{token_length}"
            results[input_key]['experiments'][token_key] = {}

    total_experiments = len(input_texts) * len(token_lengths) * len(model_names)
    current_experiment = 0

    print(f"Starting sweep: {len(model_names)} models × {len(input_texts)} inputs × {len(token_lengths)} token lengths = {total_experiments} experiments")
    print("="*60)

    # Process one model at a time
    for model_name in model_names:
        print(f"\n{'='*60}")
        print(f"Loading model: {model_name}")
        print(f"{'='*60}")

        try:
            # Clear CUDA cache before loading model
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

            # Load the model once
            generator = pipeline(
                'text-generation',
                model=model_name,
                device_map='auto' if torch.cuda.is_available() else None,
                torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
            )

            print(f"✓ Model loaded successfully")

            # Test all input/token combinations with this model
            for i, input_text in enumerate(input_texts):
                input_key = f"input_{i}"

                for token_length in token_lengths:
                    current_experiment += 1
                    token_key = f"token_length_{token_length}"

                    print(f"\n[{current_experiment}/{total_experiments}] Input {i}, Tokens={token_length}")
                    print(f"Input: '{input_text[:50]}...' " if len(input_text) > 50 else f"Input: '{input_text}'")

                    try:
                        # Generate text
                        output = generator(
                            input_text,
                            max_new_tokens=token_length,
                            num_return_sequences=1,
                            truncation=True,
                            return_full_text=False,
                            pad_token_id=generator.tokenizer.eos_token_id,
                            do_sample=True,
                            temperature=0.7
                        )

                        generated_text = output[0]['generated_text']
                        results[input_key]['experiments'][token_key][model_name] = generated_text
                        print(f"  ✓ Generated {len(generated_text)} chars")

                    except Exception as e:
                        error_msg = str(e)
                        if torch.cuda.is_available() and "CUDA" in error_msg:
                            torch.cuda.synchronize()
                        results[input_key]['experiments'][token_key][model_name] = f"Error: {error_msg}"
                        print(f"  ✗ Error: {error_msg}")

            # Clean up model after testing all combinations
            del generator
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
            print(f"\n✓ Completed all tests for {model_name}")

        except Exception as e:
            error_msg = str(e)
            if torch.cuda.is_available() and "CUDA" in error_msg:
                torch.cuda.synchronize()
            print(f"\n✗ Failed to load model {model_name}: {error_msg}")

            # Mark all experiments for this model as failed
            for i in range(len(input_texts)):
                input_key = f"input_{i}"
                for token_length in token_lengths:
                    token_key = f"token_length_{token_length}"
                    results[input_key]['experiments'][token_key][model_name] = f"Error: Model load failed - {error_msg}"

    print("\n" + "="*60)
    print("✓ Sweep complete!")
    print("="*60)

    return results


def save_results(results: Dict[str, Any], filename: str = "experiment_results.json"):
    """
    Save experiment results to a JSON file.

    Args:
        results: The results dictionary from sweep_experiments
        filename: Output filename (default: 'experiment_results.json')
    """
    with open(filename, 'w') as f:
        json.dump(results, f, indent=2)
    print(f"\n✓ Results saved to {filename}")


def print_summary(results: Dict[str, Any]):
    """
    Print a summary of the experiment results.

    Args:
        results: The results dictionary from sweep_experiments
    """
    print("\n" + "="*60)
    print("EXPERIMENT SUMMARY")
    print("="*60)

    for input_key, input_data in results.items():
        print(f"\n{input_key}: '{input_data['text'][:50]}...'")
        for token_key, model_outputs in input_data['experiments'].items():
            print(f"  {token_key}:")
            for model_name in model_outputs.keys():
                output_preview = str(model_outputs[model_name])[:80]
                status = "✓" if not output_preview.startswith("Error:") else "✗"
                print(f"    {status} {model_name}: {output_preview}...")


# Example usage
if __name__ == "__main__":
    models = [
        #'gpt2',
        #'distilgpt2',
        'EleutherAI/gpt-neo-125M',
        'facebook/opt-125m',
        'cerebras/Cerebras-GPT-111M',
        #'gpt2-medium',
        #'gpt2-large',
        'EleutherAI/gpt-neo-1.3B',
        'facebook/opt-1.3b',
        'EleutherAI/pythia-1.4b'
    ]

    inputs = [
        sample_corpus_shortened,
        sample_corpus_summary,
        sample_corpus_shortened_no_lead,
        sample_corpus_summary_no_lead
    ]

    token_limits = [5, 10, 15, 20]

    # Run the sweep
    if False:
      all_results = sweep_experiments(models, inputs, token_limits)

      # Print summary
      print_summary(all_results)

      # Save to file
      save_results(all_results)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from transformers import AutoModelForCausalLM, AutoTokenizer
from captum.attr import (
    IntegratedGradients,
    LayerIntegratedGradients,
    DeepLift,
    LayerDeepLift,
    GradientShap,
    Saliency,
    InputXGradient
)
from typing import Tuple, List, Dict
import warnings
warnings.filterwarnings('ignore')


def cycle_interp_methods(chosen_input: str, chosen_model: str, chosen_token_length: int):
    """
    Cycles through multiple interpretability methods from Captum library
    and visualizes token importance for text generation.

    Args:
        chosen_input: Input text to analyze
        chosen_model: HuggingFace model name
        chosen_token_length: Number of tokens to generate
    """

    print(f"Loading model: {chosen_model}")
    print(f"Input: {chosen_input}")
    print(f"Token length: {chosen_token_length}")
    print("="*80)

    # Load model and tokenizer with fallback options
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    try:
        # Try loading tokenizer from the model repo
        tokenizer = AutoTokenizer.from_pretrained(chosen_model, trust_remote_code=True)
    except Exception as e:
        print(f"Could not load tokenizer from {chosen_model}, trying fallback...")
        # Common fallback tokenizers for different model families
        fallback_tokenizers = {
            'cerebras': 'gpt2',
            'EleutherAI': 'gpt2',
            'facebook/opt': 'facebook/opt-125m',
            'bigscience/bloom': 'bigscience/bloom-560m'
        }

        # Find appropriate fallback
        tokenizer_name = 'gpt2'  # Default fallback
        for key, fallback in fallback_tokenizers.items():
            if key in chosen_model.lower():
                tokenizer_name = fallback
                break

        print(f"Using fallback tokenizer: {tokenizer_name}")
        tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

    # Set padding token if not available
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    model = AutoModelForCausalLM.from_pretrained(
        chosen_model,
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
        trust_remote_code=True
    ).to(device)
    model.eval()

    # Tokenize input
    inputs = tokenizer(chosen_input, return_tensors='pt').to(device)
    input_ids = inputs['input_ids']

    # Get embeddings layer for proper attribution
    embedding_layer = model.get_input_embeddings()
    input_embeddings = embedding_layer(input_ids).detach()
    input_embeddings.requires_grad = True

    # Generate output to analyze
    with torch.no_grad():
        outputs = model.generate(
            input_ids,
            max_new_tokens=chosen_token_length,
            do_sample=True,
            temperature=0.01,
            pad_token_id=tokenizer.eos_token_id,
            return_dict_in_generate=True,
            output_scores=True
        )

    generated_ids = outputs.sequences[0]
    generated_text = tokenizer.decode(generated_ids, skip_special_tokens=True)
    tokens = tokenizer.convert_ids_to_tokens(input_ids[0])

    print(f"\nGenerated text: {generated_text}")
    print(f"\nInput tokens ({len(tokens)}): {tokens}")

    # Define forward function for Captum that works with embeddings
    def forward_func(embeddings):
        """Forward function that takes embeddings and returns logits"""
        # Get output from model using embeddings directly
        outputs = model(inputs_embeds=embeddings)
        # Return logits for the last token
        return outputs.logits[:, -1, :]

    # Wrapper to convert token IDs to embeddings
    def predict_from_ids(input_ids):
        """Convert input IDs to embeddings and run forward"""
        embeddings = embedding_layer(input_ids.long())
        return forward_func(embeddings)

    # Dictionary to store attributions
    attributions_dict = {}


    # Show generated tokens
    generated_tokens = tokenizer.convert_ids_to_tokens(generated_ids)
    num_input_tokens = len(input_ids[0])
    output_tokens = generated_tokens[num_input_tokens:]

    print(f"\n" + "="*80)
    print("GENERATED OUTPUT TOKENS:")
    print("="*80)
    for idx, token in enumerate(output_tokens):
        print(f"  [{idx}] {token}")
    print("="*80)

    # Get user selection
    while True:
        try:
            user_input = input(f"\nSelect token to explain (0-{len(output_tokens)-1}): ")
            selected_idx = int(user_input)
            if 0 <= selected_idx < len(output_tokens):
                break
            else:
                print(f"Please enter a number between 0 and {len(output_tokens)-1}")
        except ValueError:
            print("Please enter a valid number")
        except (KeyboardInterrupt, EOFError):
            print("\nUsing default: token 0")
            selected_idx = 0
            break

    # Get the actual token ID from the selected position
    target_token_id = generated_ids[num_input_tokens + selected_idx].item()
    selected_token = output_tokens[selected_idx]

    print(f"\n" + "="*80)
    print(f"Analyzing importance for token [{selected_idx}]: '{selected_token}'")
    print(f"Token ID: {target_token_id}")
    print("="*80)

    # Method 1: Integrated Gradients
    print("\n1. Computing Integrated Gradients...")
    try:
        ig = IntegratedGradients(forward_func)
        baseline = torch.zeros_like(input_embeddings)
        attributions_ig = ig.attribute(
            input_embeddings,
            baselines=baseline,
            target=target_token_id,
            n_steps=50
        )
        # Sum across embedding dimension to get token-level attributions
        attributions_dict['Integrated Gradients'] = attributions_ig.sum(dim=2).squeeze(0).cpu().detach().numpy()
        print("   ✓ Complete")
    except Exception as e:
        print(f"   ✗ Error: {e}")

    # Method 2: Saliency (using embeddings)
    print("\n2. Computing Saliency...")
    try:
        saliency = Saliency(forward_func)
        attributions_sal = saliency.attribute(
            input_embeddings,
            target=target_token_id
        )
        attributions_dict['Saliency'] = attributions_sal.sum(dim=2).squeeze(0).cpu().detach().numpy()
        print("   ✓ Complete")
    except Exception as e:
        print(f"   ✗ Error: {e}")

    # Method 3: DeepLift (using embedding-based approach)
    print("\n3. Computing DeepLift...")
    try:
        # Use DeepLift with embeddings instead of LayerDeepLift
        dl = DeepLift(forward_func)
        baseline = torch.zeros_like(input_embeddings)
        attributions_dl = dl.attribute(
            input_embeddings,
            baselines=baseline,
            target=target_token_id
        )
        attributions_dict['DeepLift'] = attributions_dl.sum(dim=2).squeeze(0).cpu().detach().numpy()
        print("   ✓ Complete")
    except Exception as e:
        print(f"   ✗ Error: {e}")

    # Method 4: GradientShap
    print("\n4. Computing GradientShap...")
    try:
        gs = GradientShap(forward_func)
        # Create baseline distribution with embeddings
        baseline_dist = torch.cat([torch.zeros_like(input_embeddings) for _ in range(5)])
        attributions_gs = gs.attribute(
            input_embeddings,
            baselines=baseline_dist,
            target=target_token_id,
            n_samples=10
        )
        attributions_dict['GradientShap'] = attributions_gs.sum(dim=2).squeeze(0).cpu().detach().numpy()
        print("   ✓ Complete")
    except Exception as e:
        print(f"   ✗ Error: {e}")

    # Method 5: Input X Gradient (simpler alternative)
    print("\n5. Computing Input X Gradient...")
    try:
        from captum.attr import InputXGradient
        ixg = InputXGradient(forward_func)
        attributions_ixg = ixg.attribute(
            input_embeddings,
            target=target_token_id
        )
        attributions_dict['Input X Gradient'] = attributions_ixg.sum(dim=2).squeeze(0).cpu().detach().numpy()
        print("   ✓ Complete")
    except Exception as e:
        print(f"   ✗ Error: {e}")

    # Visualize results
    print("\n" + "="*80)
    print("Generating visualizations...")
    print("="*80)

    visualize_attributions(tokens[:len(input_ids[0])], attributions_dict, chosen_input, target_token_id)
    print_top_tokens(tokens[:len(input_ids[0])], attributions_dict)

    # Clean up
    del model
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    return attributions_dict


def print_top_tokens(tokens: List[str], attributions_dict: Dict[str, np.ndarray], top_k: int = 10):
    """
    Print the top K most important tokens for each interpretability method.

    Args:
        tokens: List of token strings
        attributions_dict: Dictionary mapping method names to attribution arrays
        top_k: Number of top tokens to display (default: 10)
    """

    if len(attributions_dict) == 0:
        print("No attributions available!")
        return

    print("\n" + "="*80)
    print(f"TOP {top_k} MOST IMPORTANT TOKENS BY METHOD")
    print("="*80)

    for method_name, attributions in attributions_dict.items():
        # Get absolute values and normalize
        attr_abs = np.abs(attributions)

        # Get top K indices
        actual_k = min(top_k, len(tokens))
        top_indices = np.argsort(attr_abs)[-actual_k:][::-1]

        print(f"\n{method_name}:")
        print("-" * 60)

        for rank, idx in enumerate(top_indices, 1):
            token = tokens[idx]
            score = attr_abs[idx]
            # Normalize to percentage of max
            pct = (score / (np.max(attr_abs) + 1e-10)) * 100

            # Create visual bar
            bar_length = int(pct / 5)  # Scale to 20 chars max
            bar = "█" * bar_length

            print(f"  {rank:2d}. [{idx:2d}] {token:15s} | {bar:20s} {pct:5.1f}%")

    print("\n" + "="*80)

def visualize_attributions(tokens: List[str], attributions_dict: Dict[str, np.ndarray],
                          input_text: str, target_token: str):
    """
    Create matplotlib visualizations for token attributions from different methods.

    Args:
        tokens: List of token strings
        attributions_dict: Dictionary mapping method names to attribution arrays
        input_text: Original input text for title
        target_token: The target token being explained
    """

    num_methods = len(attributions_dict)
    if num_methods == 0:
        print("No attributions to visualize!")
        return

    # Create figure with subplots
    fig, axes = plt.subplots(num_methods, 2, figsize=(16, 4*num_methods))
    if num_methods == 1:
        axes = axes.reshape(1, -1)

    fig.suptitle(f'Token Importance Analysis - Explaining: "{target_token}"\nInput: "{input_text[:60]}..."',
                 fontsize=14, fontweight='bold')

    for idx, (method_name, attributions) in enumerate(attributions_dict.items()):
        # Normalize attributions
        attr_norm = np.abs(attributions)
        attr_norm = attr_norm / (np.max(attr_norm) + 1e-10)

        # Plot 1: Bar chart
        ax1 = axes[idx, 0]
        colors = plt.cm.RdYlGn(attr_norm)
        bars = ax1.bar(range(len(tokens)), attr_norm, color=colors)
        ax1.set_xlabel('Token Position', fontsize=11)
        ax1.set_ylabel('Attribution Magnitude', fontsize=11)
        ax1.set_title(f'{method_name} - Bar Chart', fontsize=12, fontweight='bold')
        ax1.set_xticks(range(len(tokens)))
        ax1.set_xticklabels(tokens, rotation=45, ha='right', fontsize=9)
        ax1.grid(axis='y', alpha=0.3)

        # Plot 2: Heatmap
        ax2 = axes[idx, 1]
        attr_2d = attr_norm.reshape(1, -1)
        im = ax2.imshow(attr_2d, cmap='RdYlGn', aspect='auto', vmin=0, vmax=1)
        ax2.set_xlabel('Token Position', fontsize=11)
        ax2.set_title(f'{method_name} - Heatmap', fontsize=12, fontweight='bold')
        ax2.set_xticks(range(len(tokens)))
        ax2.set_xticklabels(tokens, rotation=45, ha='right', fontsize=9)
        ax2.set_yticks([])

        # Add colorbar
        cbar = plt.colorbar(im, ax=ax2, fraction=0.046, pad=0.04)
        cbar.set_label('Importance', fontsize=10)

        # Add text annotations for top 3 most important tokens
        top_3_idx = np.argsort(attr_norm)[-3:][::-1]
        for i, token_idx in enumerate(top_3_idx):
            ax1.text(token_idx, attr_norm[token_idx], f'#{i+1}',
                    ha='center', va='bottom', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Create comparison plot
    create_comparison_plot(tokens, attributions_dict)


def create_comparison_plot(tokens: List[str], attributions_dict: Dict[str, np.ndarray]):
    """
    Create a comparison plot showing all methods side by side.

    Args:
        tokens: List of token strings
        attributions_dict: Dictionary mapping method names to attribution arrays
    """

    fig, ax = plt.subplots(figsize=(14, 6))

    x = np.arange(len(tokens))
    width = 0.8 / len(attributions_dict)

    for idx, (method_name, attributions) in enumerate(attributions_dict.items()):
        attr_norm = np.abs(attributions)
        attr_norm = attr_norm / (np.max(attr_norm) + 1e-10)

        offset = (idx - len(attributions_dict)/2) * width + width/2
        ax.bar(x + offset, attr_norm, width, label=method_name, alpha=0.8)

    ax.set_xlabel('Tokens', fontsize=12, fontweight='bold')
    ax.set_ylabel('Normalized Attribution', fontsize=12, fontweight='bold')
    ax.set_title('Comparison of Interpretability Methods', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(tokens, rotation=45, ha='right')
    ax.legend(loc='upper right', fontsize=10)
    ax.grid(axis='y', alpha=0.3)

    plt.tight_layout()
    plt.show()


# Example usage
if __name__ == "__main__":
    # Example call after running sweep experiments
    cycle_interp_methods(
        chosen_input=chosen_sample_corpus,
        chosen_model=chosen_model,
        chosen_token_length=chosen_token_length
    )
```
